```python
def click(event):
    global scvaluess
    text = event.widget.cget('text')
    print(text)
    if text == '=':
        if scvalue.get().isdigit():
            value = int(scvalue.get())
        else:
            try:
                value = eval(screen.get())
            except Exception as e:
                print(e)
                value = 'Error'
                scvalue.set(value)
                screen.update()
        
    elif text == 'c':
        scvalue.set('')
        screen.update()
    else:
        scvalue.set(scvalue.get() + text)
        screen.update()
root = Tk()
root.geometry('300x556')
root.title('MyCalculator')
root.wm_iconbitmap('475497.png')
        
scvalue = StringVar('')
scvalue.set('')
screen = Entry(root, textvar = scvalue, font='lucida 40 bold')
screen.pack(fill=X, ipadx=8, pady=10, padx=10)

f = Frame(root, bg = 'grey')
b = Button(f, text='7', padx=10, pady=10, font = 'lucida 20 bold')
b.pack(side=LEFT, padx=1, pady=2)
b.bind('<Button-1>', click)

b = Button(f, text='8',  padx=10, pady=10, font = 'lucida 20 bold')
b.pack(side=LEFT, padx=1, pady=2)
b.bind('<Button-1>', click)

b = Button(f, text='9',  padx=10, pady=10, font = 'lucida 20 bold')
b.pack(side=LEFT, padx=1, pady=2)
b.bind('<Button-1>', click)

b = Button(f, text='+', padx=10, pady=10, font = 'lucida 20 bold')
b.pack(side=LEFT, padx=1, pady=2)
b.bind('<Button-1>', click)
f.pack()

f = Frame(root, bg = 'grey')
b = Button(f, text='6', padx=10, pady=10, font = 'lucida 20 bold')
b.pack(side=LEFT, padx=1, pady=2)
b.bind('<Button-1>', click)

b = Button(f, text='5', padx=10, pady=10, font = 'lucida 20 bold')
b.pack(side=LEFT, anchor='ne', padx=1, pady=2)
b.bind('<Button-1>', click)

b = Button(f, text='4', padx=10, pady=10, font = 'lucida 20 bold')
b.pack(side=LEFT, padx=1, pady=2)
b.bind('<Button-1>', click)

b = Button(f, text='c', padx=10, pady=10, font = 'lucida 20 bold')
b.pack(side=LEFT, padx=1, pady=2)
b.bind('<Button-1>', click)

f.pack()

f = Frame(root, bg = 'grey')
b = Button(f, text='3', padx=11, pady=10, font = 'lucida 20 bold')
b.pack(side=LEFT, padx=1, pady=2)
b.bind('<Button-1>', click)

b = Button(f, text='2', padx=10, pady=10, font = 'lucida 20 bold')
b.pack(side=LEFT, padx=1, pady=2)
b.bind('<Button-1>', click)

b = Button(f, text='1', padx=10, pady=10, font = 'lucida 20 bold')
b.pack(side=LEFT, padx=1, pady=2)
b.bind('<Button-1>', click)

b = Button(f, text='/', padx=12, pady=11, font = 'lucida 19 bold')
b.pack(side=LEFT, padx=1, pady=2)
b.bind('<Button-1>', click)
f.pack()

f = Frame(root, bg = 'grey')
b = Button(f, text='0', padx=11, pady=11, font = 'lucida 19 bold')
b.pack(side=LEFT, padx=1, pady=2)
b.bind('<Button-1>', click)

b = Button(f, text='-', padx=12, pady=10, font = 'lucida 20 bold')
b.pack(side=LEFT,padx=1, pady=2)
b.bind('<Button-1>', click)

b = Button(f, text='x', padx=11, pady=10, font = 'lucida 20 bold')
b.pack(side=LEFT, padx=1, pady=2)
b.bind('<Button-1>', click)

b = Button(f, text='%', padx=10, pady=10, font = 'lucida 19 bold')
b.pack(side=LEFT, padx=1, pady=1)
b.bind('<Button-1>', click)
f.pack()

f = Frame(root, bg = 'grey')
b = Button(f, text='=', padx=10, pady=10, font = 'lucida 19 bold')
b.pack(side=LEFT, padx=1, pady=1)
b.bind('<Button-1>', click)

b = Button(f, text='.', padx=10, pady=10, font = 'lucida 19 bold')
b.pack(side=LEFT, padx=1, pady=1)
b.bind('<Button-1>', click)

b = Button(f, text='00', padx=10, pady=10, font = 'lucida 19 bold')
b.pack(side=LEFT, padx=1, pady=1)
b.bind('<Button-1>', click)
f.pack()

root.mainloop()
```
